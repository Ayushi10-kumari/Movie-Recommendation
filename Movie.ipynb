```python
movies=movie[['id','title','overview','genre']]
```

```python
movies
```

```python
movies.loc[:, 'tags'] = movies['overview'] + movies['genre']
```

```python
movies
```

```python
print(movies.columns)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
movies[movies['title']=="The Godfather"].index[0]
```

```python
dis=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in dis[0:5]:
    print(movies.iloc[i[0]].title)
```

```python
def recommand(movie_title):
    # Assuming 'movies' is the DataFrame and 'similarity' is a precomputed matrix
    index = movies[movies['title'] == movie_title].index[0]
    dis = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    
    # Printing the top 5 recommendations
    for i in dis[1:6]:  # Start from 1 to skip the movie itself
        print(movies.iloc[i[0]]['title'])

    
```

```python
recommand("Iron Man")
```
